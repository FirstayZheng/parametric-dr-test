```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt("/home/amax/parametric-dr/Face/face.csv", delimiter=',')
data.shape
```

```python
def normalization(data):
    # 灰度值的范围是 0---255， 这里|做一个归一化
    _range = 255
    return (data - 0) / _range
normed_data = normalization(data)
```

```python
from sklearn.manifold import MDS

embedding = MDS(n_components=2)
X_transformed = embedding.fit_transform(normed_data)
plt.scatter(X_transformed[:,0], X_transformed[:,1])
plt.savefig('./face_MDS_normed.png')
plt.show()

X_transformed = embedding.fit_transform(data)
plt.scatter(X_transformed[:,0], X_transformed[:,1])
plt.savefig('./face_MDS.png')
```

```python
from sklearn.decomposition import PCA

pca_data= PCA(n_components=50).fit_transform(normed_data)
print(pca_data.shape)
embedding = MDS(n_components=2)
X_transformed = embedding.fit_transform(normed_data)
plt.scatter(X_transformed[:,0], X_transformed[:,1])
plt.savefig('./face_MDS_pca.png')
plt.show()
```

```python
from sklearn.manifold import Isomap
from sklearn.decomposition import PCA


pca_data= PCA(n_components=50).fit_transform(normed_data)
print(pca_data.shape)
embedding = Isomap(n_components=2)
X_transformed = embedding.fit_transform(pca_data)
print(X_transformed.shape)
plt.scatter(X_transformed[:,0], X_transformed[:,1])
plt.savefig('./face_isomap.png')
plt.show()
```

```python
pca_data= PCA(n_components=2).fit_transform(normed_data)
plt.scatter(pca_data[:,0], pca_data[:,1])
plt.savefig('./face_pca.png')
plt.show()
```
